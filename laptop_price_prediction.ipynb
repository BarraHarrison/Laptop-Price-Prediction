```python
def turn_memory_into_MB(value):
    if "GB" in value:
        return float(value[:value.find("GB")]) * 1000
    elif "TB" in value:
        return float(value[:value.find("TB")]) * 1000000
```

```python
data_frame["Memory Amount"].apply(turn_memory_into_MB)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18,15))
sns.heatmap(data_frame.corr(numeric_only=True), annot=True, cmap="YlGnBu")
```

```python
target_correlations = data_frame.corr(numeric_only=True)["Price_euros"].apply(abs).sort_values()
```

```python
target_correlations
```

```python
selected_features = target_correlations[-20:].index
```

```python
selected_features
```

```python
selected_features = list(selected_features)
```

```python
selected_features
```

```python
limited_data_frame = data_frame[selected_features]
```

```python
limited_data_frame
```

```python
plt.figure(figsize=(18,15))
sns.heatmap(limited_data_frame.corr(), annot=True, cmap="YlGnBu")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x, y = limited_data_frame.drop("Price_euros", axis=1), limited_data_frame["Price_euros"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
forest = RandomForestRegressor()

forest.fit(x_train_scaled, y_train)
```

```python
forest.score(x_test_scaled, y_test)
```

```python
y_prediction = forest.predict(x_test_scaled)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_prediction, y_test)
plt.plot(range(0, 6000), range(0, 6000), c="red")
```
